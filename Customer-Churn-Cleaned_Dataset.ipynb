```python
# Import Libraries and Load Dataset.
import pandas as pd
import numpy as np
```

```python
# Checking the current working directory to verify the file path.
import os
os.getcwd()
```

```python
df=pd.read_csv("C:/Users/jishu/OneDrive/Ritu/Customer-Churn-Dataset.csv")
df.head()
```

```python
# Perform a thorough examination of the dataset to identify any missing values and duplicate records.
print("Missing values per column:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Convert 'TotalCharges' to numeric and check results
df['TotalCharges'] = df['TotalCharges'].astype(str).str.strip()  # remove spaces
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Verify conversion
print(f"TotalCharges dtype: {df['TotalCharges'].dtype}")
print(f"Null values in TotalCharges: {df['TotalCharges'].isna().sum()}")
```

```python
# Replace NaN values in 'TotalCharges' with 0
df['TotalCharges'] = df['TotalCharges'].replace(np.nan, 0)

# Verify no missing values remain
print(df.isnull().sum())
```

```python
# Remove duplicates and show summary
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

print(f"Duplicates removed: {before - after}")
print(f"New shape: {df.shape}")
```

```python
df.info()
```

```python
# Clean inconsistent categorical values
cols_to_clean = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'MultipleLines'
]

df[cols_to_clean] = df[cols_to_clean].replace({
    'No internet service': 'No',
    'No phone service': 'No'
})

# Check cleaned values
for col in cols_to_clean:
    print(f"{col} unique values: {df[col].unique()}")

```

```python
# Step 8: Export the cleaned dataset
df.to_csv("Telco_Customer_Cleaned.csv", index=False)
print("Cleaned dataset saved as Telco_Customer_Cleaned.csv")
```
